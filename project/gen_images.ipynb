```python
import os
import cv2
import shutil

def extract_first_frame(video_path, output_image_path):
    video = cv2.VideoCapture(video_path)
    
    if not video.isOpened():
        print(f"Error opening video file: {video_path}")
        return False
    
    ret, frame = video.read()
    
    if ret:
        cv2.imwrite(output_image_path, frame)
        print(f"Saved first frame from {video_path} to {output_image_path}")
        return True
    else:
        print(f"Failed to read the first frame from {video_path}")
        return False

def process_folder(input_folder, output_folder):
    for root, dirs, files in os.walk(input_folder):
        relative_path = os.path.relpath(root, input_folder)
        new_dir = os.path.join(output_folder, relative_path)
        
        if not os.path.exists(new_dir):
            os.makedirs(new_dir)
        
        for file in files:
            file_path = os.path.join(root, file)
            if file.lower().endswith(('.mp4', '.avi', '.mov', '.mkv')):
                output_image_path = os.path.join(new_dir, f"{os.path.splitext(file)[0]}.jpg")
                extract_first_frame(file_path, output_image_path)
```

```python
if __name__ == "__main__":
    input_folder = "video"
    output_folder = "images"
    
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    process_folder(input_folder, output_folder)
```
